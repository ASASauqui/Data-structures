```python
class Node:
    def __init__(self,data):
        self.next = None
        self.prev = None
        self.data = data
        
    def __str__(self):
        s = str(self.data)
        s += " prev:" + str(hex(id(self.prev)))
        s += " &:" + str(hex(id(self)))
        s += " next:" + str(hex(id(self.next))) #Direcciones en Python
        return s 
```

```python
class DoublyLinkedList:
    def __init__(self):
        self.first = None
        self.last = None
        
    def push_front(self,data):
        new_node = Node(data)
        if self.first is None:
            self.first = new_node
            self.last = new_node
        else:
            new_node.next = self.first
            self.first.prev = new_node
            self.first = new_node
            
            
    def push_back(self,data):
        new_node = Node(data)
        if self.first is None:
            self.first = new_node
            self.last = new_node
        else:
            new_node.prev = self.last
            self.last.next = new_node
            self.last = new_node
              
                
    def pop_front(self):
        if self.first is None:
            return
        if self.first is self.last:
            self.first = None
            self.last = None
            return
        self.first.next.prev = None
        self.first = self.first.next
        
    
    
    def pop_back(self):
        if self.last is None:
            return
        if self.first is self.last:
            self.first = None
            self.last = None
            return
        self.last.prev.next = None 
        self.last = self.last.prev
    
    
    def remove(self,data):
        aux = self.first
        while aux is not None:
            if aux.data == data:
                break
            aux = aux.next
        if aux is None: return      
        
        if aux.prev is not None:
            aux.prev.next = aux.next
        if aux.next is not None:
            aux.next.prev = aux.prev
        
        if aux == self.first:
            self.first = aux.next
        if aux == self.last:
            self.last = aux.prev
            
        aux.next = None
        aux.prev = None

    def print_reverse(self):
        if self.last is None:
            return "[]"
        else:
            aux = self.last
            s = ""
            while aux is not None:
                s += aux.__str__() + ", "
                aux = aux.prev
            return s
    
    def __str__(self):
        if self.first is None:
            return "[]"
        else:
            aux = self.first
            s = ""
            while aux is not None:
                s += aux.__str__() + ", "
                aux = aux.next
            return s
        
l = DoublyLinkedList()
l.push_back(3)
l.push_back(9)
l.push_back(10)
l.push_front(13)
print(l)
l.remove(9)

print()
print(l)

l.pop_back()
l.pop_front()

print()
print(l)

print()
print(l.print_reverse())
```
