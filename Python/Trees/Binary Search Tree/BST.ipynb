# STACK

```python
class NodeStack:
    def __init__(self,data):
        self.next = None
        self.data = data
    
    def __str__(self):
        s = str(self.data)
        s += " &: " + str(hex(id(self)))
        s += " next: " + str(hex(id(self.next)))
        return s 
```

```python
class Stack:
    def __init__(self):
        self.first = None
    
    def push(self,data):
        n = NodeStack(data)
        if self.first is None:
            self.first = n
        else:
            n.next = self.first
            self.first = n
            
    def pop(self):
        if self.first is None:
            return None
        n = self.first
        if self.first.next is None:
            self.first = None
        else:
            self.first = self.first.next
        return n.data
            
    def top(self):
        if self.first is None:
            return None
        else:
            return self.first.data
        
    def isEmpty(self):
        if self.first is None:
            return True
        else:
            return False
        
    def __str__(self):
        if self.first is None:
            return "[]"
        else:
            aux = self.first
            s = ""
            while aux is not None:
                s += aux.__str__() + ", "
                aux = aux.next
            return s
```

# QUEUE

```python
class NodeQueue:
    def __init__(self,data):
        self.next = None
        self.data = data
    
    def __str__(self):
        s = str(self.data)
        s += " &: " + str(hex(id(self)))
        s += " next: " + str(hex(id(self.next)))
        return s 
```

```python
class Queue:
    def __init__(self):
        self.first = None
        self.last = None
    
    def push(self,data):
        n = NodeQueue(data)
        if self.first is None:
            self.first = n
            self.last = n
        else:
            self.last.next = n
            self.last = n
            
    def pop(self):
        if self.first is None:
            return None
        n = self.first
        if self.first is self.last:
            self.first = None
            self.last = None
        else:
            self.first = self.first.next
        return n.data
            
    def top(self):
        if self.first is None:
            return None
        else:
            return self.first.data
        
    def isEmpty(self):
        if self.first is None:
            return True
        else:
            return False
        
    def __str__(self):
        if self.first is None:
            return "[]"
        else:
            aux = self.first
            s = ""
            while aux is not None:
                s += aux.__str__() + ", "
                aux = aux.next
            return s
```

# TREE

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
    def __str__(self):
        return str(self.data)
    
class BST:
    def __init__(self):
        self.root = None
    
    def insert(self,data):
        n = Node(data)
        if self.root is None:
            self.root = n
        else:
            aux = self.root
            while(True):
                if data <= aux.data: #izquierda
                    if aux.left is None:
                        aux.left = n
                        break
                    else:
                        aux = aux.left
                    
                else:                #derecha
                    if aux.right is None:
                        aux.right = n
                        break
                    else:
                        aux = aux.right
                        
    def contains(self,data):
        if self.root is None:
            return False
        aux = self.root
        while aux is not None:
            if aux.data is data:
                return True
            if data < aux.data:
                if aux.left is None:
                    return False
                aux = aux.left
            else:
                if aux.right is None:
                    return False
                aux = aux.right
        return False
        
    def __print_node(self,space,node):
        if node is None: return ""
        
        s = space + str(node.data) + "\n"
        s +=  self.__print_node(space+" l:",node.left)
        s +=  self.__print_node(space+" r:",node.right)
        return s
    
    def __str__(self):
        return self.__print_node("",self.root)
    
    def mostrarArbol(self,node,cont):
        if node is None:
            return
        self.mostrarArbol(node.right,cont+1)
        s = ""
        for i in range(cont):
            s+="  "
        print(s,node.data)
        self.mostrarArbol(node.left,cont+1)
        
    def print(self):
        self.mostrarArbol(self.root,0)
    
    def DFS(self):
        print("DFS:")
        stack = Stack()
        if self.root is None:
            print(" ")
        else:
            stack.push(self.root)
            while stack.isEmpty() is not True:
                n = stack.pop()
                print( "   "+ str(n.data) + " ")
                if n.right is not None:
                    stack.push(n.right)
                if n.left is not None:
                    stack.push(n.left)
    
    def BFS(self):
        print("BFS:")
        queue = Queue()
        if self.root is None:
            print(" ")
        else:
            queue.push(self.root)
            while queue.isEmpty() is not True:
                n = queue.pop()
                print( "   "+ str(n.data) + " ")
                if n.left is not None:
                    queue.push(n.left)
                if n.right is not None:
                    queue.push(n.right)
    
    def preorderR(self,nodo):
        if nodo is not None:
            print("   " + str(nodo.data))
            self.preorderR(nodo.left)
            self.preorderR(nodo.right)
        return ""
    
    def preorder(self):
        print("Preorder")
        if self.root is None:
            print(" ")
        else:
            self.preorderR(self.root)
    
    def inorderR(self,nodo):
        if nodo is not None:
            self.inorderR(nodo.left)
            print("   " + str(nodo.data))
            self.inorderR(nodo.right)
        return ""
    
    def inorder(self):
        print("Inorder")
        if self.root is None:
            print(" ")
        else:
            self.inorderR(self.root)
    
    def postorderR(self,nodo):
        if nodo is not None:
            self.postorderR(nodo.left)
            self.postorderR(nodo.right)
            print("   " + str(nodo.data))
        return ""
    
    def postorder(self):
        print("Postorder")
        if self.root is None:
            print(" ")
        else:
            self.postorderR(self.root)
    
    def __search_parent_node(self,data):
        if self.root is None:
            return None, None
        parent = None
        node = self.root
        while True:
            if data == node.data:
                return parent, node
            elif data < node.data: #izquierda
                parent = node
                node = node.left
            elif data > node.data: #derecha
                parent = node
                node = node.right
            if node is None:
                return None,None
    
    def __remove(self,parent,node):
        if node.left is None and node.right is None:
            print('   Caso 1')
            if parent is None: self.root = None
            elif parent.left == node:  parent.left = None
            elif parent.right == node: parent.right = None
            
        elif (node.left is None and node.right is not None) or (node.left is not None and node.right is None):
            print('   Caso 2')
            if parent is None:
                if node.left is not None: self.root = node.left
                elif node.right is not None: self.root = node.right
            elif parent.left == node:
                if node.left is not None: parent.left = node.left
                elif node.right is not None: parent.left = node.right
            elif parent.right == node:
                if node.left is not None: parent.right = node.left
                elif node.right is not None: parent.right = node.right 
            
        elif node.left is not None and node.right is not None:
            print('   Caso 3')
            parent2 = node
            node2 = node.left
            while node2.right is not None:
                parent2 = node2
                node2 = node2.right
            print('   parent2:',parent2,'node2:',node2)
            self.__remove(parent2,node2)
            node2.left = node.left
            node2.right = node.right
            if parent is None: self.root = node2
            elif parent.left == node:  parent.left = node2
            elif parent.right == node: parent.right = node2
        
        
    def remove(self,data):
        print('Remove:')
        parent,node = self.__search_parent_node(data)
        if node is None:
            return
        print('   parent:',parent,'node:',node)
        self.__remove(parent,node)
        print('   Delete',node)
```
