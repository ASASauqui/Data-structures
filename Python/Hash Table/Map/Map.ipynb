```python
class Node:
    def __init__(self,data):
        self.next = None
        self.data = data
        
    def __str__(self):
        s = "("
        s += str(self.data[0]) + ", " + str(self.data[1])
        s += ")"
        return s 
```

```python
class SinglyLinkedList:
    def __init__(self):
        self.head = None
        
    def insert_tail(self,data):
        n = Node(data)
        if self.head is None:
            self.head = n
            #print(self.head)
        else:
            aux = self.head
            if aux.data[0] == data[0]:
                aux.data = (aux.data[0],data[1])
                return
            while aux.next is not None:
                if aux.data[0] == data[0]:
                    aux.data = (aux.data[0],data[1])
                    return
                aux = aux.next
            if aux.data[0] == data[0]:
                aux.data = (aux.data[0],data[1])
                return
            aux.next = n
            
    def remove(self,key):
        aux = self.head
        if self.head.data[0] == key: 
            self.head = self.head.next
            return
        
        while aux.next is not None:  
            if aux.next.data[0] == key:
                break
            aux = aux.next
        
        if aux.next is None:
            return
        aux.next = aux.next.next
        
    def contains(self,key):
        if self.head is None:
            return False
        if self.head.data[0] == key:
            return True
        aux = self.head
        while aux.next is not None:
            if aux.data[0] == key:
                return True
            aux = aux.next
        if aux.data[0] == key:
            return True
        return False
    
    def findValue(self,key):
        if self.head is None:
            return "[]"
        if self.head.data[0] == key:
            return self.head.data[1]
        aux = self.head
        while aux.next is not None:
            if aux.data[0] == key:
                return aux.data[1]
            aux = aux.next
        if aux.data[0] == key:
            return aux.data[1]
        return "[]"
        
            
    def __str__(self):
        if self.head is None:
            return "[]"
        else:
            aux = self.head
            s = ""
            while aux is not None:
                s += aux.__str__() + ", "
                aux = aux.next
            return s
```

```python
class Map:
    def __init__(self):
        self.tabla = (SinglyLinkedList(),SinglyLinkedList(),SinglyLinkedList(),SinglyLinkedList(),SinglyLinkedList(),SinglyLinkedList(),SinglyLinkedList())
    
    def is_number(self, n):
       try:
           float(n)
       except ValueError:
           return False
       return True
    
    def insert(self,key,value):
        if self.is_number(key) is True:
            self.tabla[int(key)%len(self.tabla)].insert_tail((key,value))
        else:
            n = 0
            for i in range(0,len(key)):
                n = n + ord(key[i])*pow(26,i)
                #print(n)
            self.tabla[n%len(self.tabla)].insert_tail((key,value))
            
    def remove(self,key):
        if self.is_number(key) is True:
            self.tabla[int(key)%len(self.tabla)].remove(key)
        else:
            n = 0
            for i in range(0,len(key)):
                n = n + ord(key[i])*pow(26,i)
            self.tabla[n%len(self.tabla)].remove(key)
            
    def contains(self,key):
        #print(self.tabla[key%len(self.tabla)].contains(key))
        if self.is_number(key) is True:
            if self.tabla[int(key)%len(self.tabla)].contains(key) is True:
                return True
        else:
            n = 0
            for i in range(0,len(key)):
                n = n + ord(key[i])*pow(26,i)
            if self.tabla[n%len(self.tabla)].contains(key) is True:
                return True
        return False
    
    def get(self,key):
        if self.contains(key) is True:
            if self.is_number(key) is True:
                return self.tabla[int(key)%len(self.tabla)].findValue(key)
            else:
                n = 0
                for i in range(0,len(key)):
                    n = n + ord(key[i])*pow(26,i)
                    #print(n)
                return self.tabla[n%len(self.tabla)].findValue(key)
        else:
            return "[]"
            
    def __str__(self):
        s = ""
        for i in range(0,len(self.tabla)):
            s += "Lista " + str(i) + ": "
            s += self.tabla[i].__str__()
            s += '\n'
        return s
```
